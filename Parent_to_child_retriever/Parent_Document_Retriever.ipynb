### **Data Ingestion**

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loaders = [
    PyPDFLoader("/content/harry_potter.pdf"),
    PyPDFLoader("/content/the_secret.pdf"),
]
```

```python
docs = []
```

```python
for loader in loaders:
  docs.extend(loader.load())
```

```python
print(docs)
```

```python
# Initialize Parent and Child Splitter

from langchain_text_splitters import RecursiveCharacterTextSplitter

child_splitter = RecursiveCharacterTextSplitter(chunk_size=400)
parent_splitter = RecursiveCharacterTextSplitter(chunk_size=2000)
```

```python
# The storage layer for the parent documents
store = InMemoryStore()
```

```python
!pip install langchain-google-genai
```

```python
import os
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY

from langchain_google_genai import GoogleGenerativeAIEmbeddings

gemini_embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
```

```python
from langchain_chroma import Chroma

vector_store = Chroma(collection_name="full_documents",
                      embedding_function=gemini_embeddings)
```

```python
from langchain.storage import InMemoryStore
from langchain_chroma import Chroma
```

```python
from langchain.retrievers import ParentDocumentRetriever

retriever = ParentDocumentRetriever(
    vectorstore=vector_store,
    docstore=store,
    child_splitter=child_splitter,
    parent_splitter=parent_splitter,
)
```

```python
retriever.add_documents(docs, ids=None)
```

```python
# Number of Parenet Chunk
len(list(store.yield_keys()))
```

```python
retriever_doc = retriever.invoke("Who is harry ?")
```

```python
len(retriever_doc[0].page_content)
```

```python
#Returning 2 chunks. Parent and Child

retriever_doc
```
