# **Loading DATA**

```python
from langchain.document_loaders import PyPDFLoader
```

```python
loader_harrypotter  = PyPDFLoader("harry_potter.pdf")
```

```python
documnet_harrypotter = loader_harrypotter.load()
```

```python
print(len(documnet_harrypotter))
```

```python
loader_the_secret = PyPDFLoader("the_secret.pdf")
```

```python
document_the_secret = loader_the_secret.load()
```

```python
print(len(document_the_secret))
```

# **Creating Text Splitter for Chunking**

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
```

```python
text_harrypotter = text_splitter.split_documents(documnet_harrypotter)
```

```python
text_the_secret = text_splitter.split_documents(document_the_secret)
```

```python
print(len(text_harrypotter))
```

```python
print(len(text_the_secret))
```

# **Load the Embedding Model to Convert the Data into Vector**

```python
from langchain.embeddings import HuggingFaceEmbeddings, HuggingFaceBgeEmbeddings
```

```python
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
hf_bge_embeddings = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-large-en")
```

# **Now ingest the Data into the Chroma Database**

```python
from langchain.vectorstores import Chroma
import chromadb
```

```python
harrypotter_vectorstore = Chroma.from_documents(text_harrypotter,
                                       hf_bge_embeddings,
                                       )
```

```python
the_secret_vectorstore = Chroma.from_documents(text_the_secret,
                                       hf_bge_embeddings,
                                       )
```

```python
retriever_harrypotter = harrypotter_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
retriever_the_secret = the_secret_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

# **Merging both Retriever**

# **It is also called lord of retriever(LOTR) **

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_harrypotter, retriever_the_secret])
```

```python
for chunks in lotr.get_relevant_documents("Who was the jon snow?"):
    print(chunks.page_content)
```

# **Now After understanding step by step create a pipeline for LLM**

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
)
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import ContextualCompressionRetriever
from langchain.document_transformers import LongContextReorder
```

```python
from re import search

filter = EmbeddingsRedundantFilter(embeddings=hf_bge_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr,search_kwargs={"k": 3, "include_metadata": True}
)
```

```python
docs = compression_retriever_reordered.get_relevant_documents("What is the secret?")
print(len(docs))
#
print(docs[0].page_content)
```
